```python
import numpy as np

np.random.seed(42)
```

```python
arr = np.random.rand(100, 100)
arr.shape # (100, 100)
```

```python
def manual_mean(x):
    total = 0.0
    for row in x:
        for val in row:
            total += val
    return total / x.size
```

```python
%timeit manual_mean(arr)
%timeit np.mean(arr)
```
